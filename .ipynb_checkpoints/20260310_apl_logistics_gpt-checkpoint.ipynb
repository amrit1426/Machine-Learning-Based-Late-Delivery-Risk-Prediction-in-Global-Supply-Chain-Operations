```python
# ============================================
# 📦 IMPORT LIBRARIES
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
warnings.filterwarnings("ignore")

# ============================================
# 📂 LOAD DATA
# ============================================

df = pd.read_csv("APL_Logistics.csv", encoding="cp1252")

print("Shape:", df.shape)
df.info()

# ============================================
# ❗ MISSING VALUES
# ============================================

print(df.isnull().sum().sort_values(ascending=False).head())

# ============================================
# 🔁 DUPLICATES
# ============================================

print("Duplicates:", df.duplicated().sum())

# ============================================
# 🎯 TARGET ANALYSIS
# ============================================

sns.countplot(x='Late_delivery_risk', data=df)
plt.title("Target Distribution")
plt.show()

# ============================================
# 📦 CORE EDA
# ============================================

df['shipping_gap'] = df['Days for shipping (real)'] - df['Days for shipment (scheduled)']

sns.histplot(df['shipping_gap'], bins=50)
plt.title("Shipping Gap Distribution")
plt.show()

sns.boxplot(x='Late_delivery_risk', y='Sales', data=df)
plt.title("Sales vs Delay")
plt.show()

# ============================================
# 🧹 DATA CLEANING
# ============================================

df_clean = df.copy()

# Numeric → median
for col in df_clean.select_dtypes(include=np.number):
    df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Categorical → Unknown
for col in df_clean.select_dtypes(include='object'):
    df_clean[col].fillna("Unknown", inplace=True)

print("Remaining missing:", df_clean.isnull().sum().sum())
```

```python
# ============================================
# ⚙️ FEATURE ENGINEERING
# ============================================

df_fe = df_clean.copy()

# 🚫 REMOVE LEAKAGE
leakage_cols = [
    'Delivery Status', 'Order Status',
    'Days for shipping (real)', 'shipping_gap',
    'Order Profit Per Order', 'Order Item Profit Ratio'
]

df_fe.drop(columns=leakage_cols, inplace=True, errors='ignore')

# 🚚 SHIPPING FEATURES
df_fe['shipping_pressure'] = df_fe['Order Item Quantity'] / (df_fe['Days for shipment (scheduled)'] + 1)

# 📦 ORDER COMPLEXITY
df_fe['high_quantity_flag'] = (
    df_fe['Order Item Quantity'] > df_fe['Order Item Quantity'].median()
).astype(int)

# 💰 FINANCIAL
df_fe['discount_pressure'] = df_fe['Order Item Discount Rate']

# 🌍 REGION RISK
region_risk = df_fe.groupby('Order Region')['Late_delivery_risk'].mean()
df_fe['region_risk_score'] = df_fe['Order Region'].map(region_risk)

# 🚚 SHIPPING MODE
df_fe['is_express'] = df_fe['Shipping Mode'].apply(lambda x: 1 if 'Express' in str(x) else 0)

# 🧹 DROP HIGH CARDINALITY
drop_cols = [
    'Customer Fname','Customer Lname','Customer Street',
    'Customer Zipcode','Product Name',
    'Customer Id','Order Customer Id',
    'Latitude','Longitude'
]

df_fe.drop(columns=drop_cols, inplace=True, errors='ignore')

# 🎯 TARGET SPLIT
y = df_fe['Late_delivery_risk']
X = df_fe.drop(columns=['Late_delivery_risk'])

# 🎯 CONTROLLED ENCODING
cat_cols = ['Shipping Mode','Customer Segment','Category Name','Type']
cat_cols = [c for c in cat_cols if c in X.columns]

X = pd.get_dummies(X, columns=cat_cols, drop_first=True)

# 🚫 DROP RAW REGION
X.drop(columns=['Order Region','Market'], inplace=True, errors='ignore')

# 🚨 ENSURE NO OBJECTS
X = X.select_dtypes(exclude='object')

# 📏 SCALING
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X[X.columns] = scaler.fit_transform(X)

print("Final Shape:", X.shape)

# Export model features for SHAP analysis
X.to_csv("data/model_features.csv", index=False)
```

```python
# ============================================
# 📦 IMPORT MODELS
# ============================================

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, f1_score

# ============================================
# 🔀 TRAIN TEST SPLIT
# ============================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# ============================================
# 🧠 TRAIN MODELS
# ============================================

log_model = LogisticRegression(class_weight='balanced', max_iter=1000)
rf_model = RandomForestClassifier(n_estimators=300, max_depth=12, class_weight='balanced', random_state=42)
xgb_model = XGBClassifier(n_estimators=200, max_depth=6, learning_rate=0.1, random_state=42)

log_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)

# ============================================
# 📊 EVALUATION FUNCTION
# ============================================

def evaluate_model(model, name, threshold=0.5):
    
    y_prob = model.predict_proba(X_test)[:,1]
    y_pred = (y_prob >= threshold).astype(int)

    print(f"\n=== {name} (Threshold={threshold}) ===")
    print("\nF1 Score:", f1_score(y_test, y_pred))
    print("\nROC-AUC:", roc_auc_score(y_test, y_prob))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

# ============================================
# 📊 EVALUATE (CRITICAL)
# ============================================

# Baseline
evaluate_model(log_model, "Logistic", 0.5)
evaluate_model(rf_model, "Random Forest", 0.5)

# Business threshold
evaluate_model(rf_model, "Random Forest", 0.35)

# Advanced model
evaluate_model(xgb_model, "XGBoost", 0.35)
```

```python
# ============================================
# 🎯 FINAL RISK PROBABILITY
# ============================================

final_threshold = 0.35

y_prob_full = rf_model.predict_proba(X)[:,1]

df_fe['risk_probability'] = y_prob_full
df_fe['risk_flag'] = (y_prob_full >= final_threshold).astype(int)

# ============================================
# 🏷️ RISK CATEGORY
# ============================================

df_fe['risk_category'] = pd.cut(
    df_fe['risk_probability'],
    bins=[0,0.3,0.6,1],
    labels=['Low','Medium','High']
)

# ============================================
# 📊 KPI METRICS
# ============================================

print("Average Risk:", df_fe['risk_probability'].mean())
print("High Risk Orders:", (df_fe['risk_category']=="High").sum())
print("High Risk %:", (df_fe['risk_category']=="High").mean()*100)

# ============================================
# 🧠 FEATURE IMPORTANCE
# ============================================

importance = pd.Series(rf_model.feature_importances_, index=X.columns)
print("\nTop Drivers:\n", importance.sort_values(ascending=False).head(10))
```

```python
# ============================================
# 💾 SAVE MODEL + SCALER
# ============================================

import joblib

os.makedirs("models", exist_ok=True)

joblib.dump(rf_model, "models/risk_model.pkl")
joblib.dump(scaler, "models/scaler.pkl")

print("✅ Model saved")
```

```python
# ============================================
# 📊 FINAL DATASET (DASHBOARD-READY)
# ============================================

os.makedirs("data", exist_ok=True)

# STEP 1: Start from ORIGINAL CLEAN DATA (NOT df_fe)
df_final = df_clean.copy()

# STEP 2: Add model predictions (aligned by index)
df_final["Late_Delivery_Probability"] = y_prob_full

# STEP 3: Risk category (PRD naming)
df_final["Risk_Category"] = pd.cut(
    df_final["Late_Delivery_Probability"],
    bins=[0, 0.3, 0.6, 1],
    labels=["Low", "Medium", "High"]
)

# STEP 4: KEEP ONLY PRD REQUIRED COLUMNS (BUT CORRECT ONES)

final_cols = [
    "Order Id",                         # REQUIRED for order-level view
    "Shipping Mode",                    # PRD: Shipping mode filter
    "Order Region",                     # PRD: Region / market selector
    "Customer Segment",                 # PRD: Customer segment filter
    "Order Item Quantity",
    "Days for shipment (scheduled)",
    "Late_Delivery_Probability",
    "Risk_Category"
]

# Keep only existing columns safely
final_cols = [c for c in final_cols if c in df_final.columns]

df_final = df_final[final_cols]

# STEP 5: SAVE
df_final.to_csv("data/processed_data.csv", index=False)

print("✅ FINAL DATASET READY (DASHBOARD COMPATIBLE)")
print(df_final.head())
```
